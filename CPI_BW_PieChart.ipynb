```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.animation as animation
from matplotlib.font_manager import FontProperties as FP
from matplotlib import colors as col

#Set ffmpeg location
plt.rcParams['animation.ffmpeg_path'] = './ffmpeg/bin/ffmpeg.exe'

#Set fig size
plt.rcParams['figure.figsize'] = (20,12)

#Set font globally
plt.rcParams['font.family'] = ['fantasy']

#Set font variable for individual object
font = FP(fname="./Font/SimHei.ttf")

#Set style -> #https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html
plt.style.use('ggplot')
```

```python
df = pd.read_excel("./Data/CPI Quatity Weight.xlsx")
df
```

```python
item = df["Type"]
item_value = df[2021]
item_value_old = df[2016]
explode = (0,0,0,0,0,0,0.2) 
```

```python
# PIE 

plt.style.use('ggplot')
plt.title('Official Consumer Price Index (CPI) basket weights ', fontsize= 26)
plt.axis('equal')
plt.pie(x=item_value.values, 
        labels=item.values,
        explode=explode, 
        autopct='%.2f%%', 
        shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.6},
        startangle=0,
        wedgeprops=dict(edgecolor='w', linewidth=1.5),
        textprops={'fontsize':14}
        )
plt.legend(loc='upper left', labels= item.values, fontsize = 14)

#Save plot as png file with transparent background and decrease the white space around the output image.
#You must save plot first before calling show(), otherwise the output image will be empty.
plt.savefig('./Output/Image/CPI_BW_PieChart.png', transparent=True, bbox_inches='tight')

plt.show()

```
